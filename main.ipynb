```python
import pandas as pd
```

```python
dataset = pd.read_csv("Titanic-Dataset.csv")
print(dataset.shape)
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
X = dataset.drop("Survived", axis=1)
y = dataset["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, stratify=y)
```

```python
dataset.isnull().sum()
```

```python
X_train['Title'] = X_train['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
X_test['Title'] = X_test['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
X_train = X_train.drop(['Name', 'Ticket', 'Cabin'], axis=1)
X_test = X_test.drop(['Name', 'Ticket', 'Cabin'], axis=1)
```

```python
X_train['FamilySize'] = X_train['SibSp'] + X_train['Parch'] + 1 
X_test['FamilySize'] = X_test['SibSp'] + X_test['Parch'] + 1
```

```python
X_train = X_train.drop(['SibSp', 'Parch'], axis=1)
X_test = X_test.drop(['SibSp', 'Parch'], axis=1)
```

```python
X_train.isnull().sum()
```

```python
combined = pd.concat([X_train, X_test])

combined['Age'] = combined.groupby('Title')['Age'].transform(lambda x: x.fillna(x.median()))

X_train['Age'] = combined.loc[X_train.index, 'Age']
X_test['Age'] = combined.loc[X_test.index, 'Age']
```

```python
X_train
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
X_train['Sex'] = le_sex.fit_transform(X_train['Sex'])
X_test['Sex'] = le_sex.transform(X_test['Sex'])
```

```python
le_embarked = LabelEncoder()
X_train['Embarked'] = le_embarked.fit_transform(X_train['Embarked'])
X_test['Embarked'] = le_embarked.transform(X_test['Embarked'])
```

```python
le_title = LabelEncoder()
X_train['Title'] = le_title.fit_transform(X_train['Title'])
X_test['Title'] = le_title.transform(X_test['Title'])
```

```python
import xgboost as xgb
from sklearn.metrics import roc_auc_score, classification_report
```

```python
model = xgb.XGBClassifier(
    objective='binary:logistic',
    eval_metric='auc',
    random_state=42
)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred_proba = model.predict_proba(X_test)[:, 1]
auc = roc_auc_score(y_test, y_pred_proba)
print(f"ROC-AUC: {auc:.4f}")

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```
